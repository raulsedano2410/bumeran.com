```python
# Librerías
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from time import sleep
import pandas as pd


options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-extensions")

# Ruta absoluta para mi pc :
driver_path = 'C:\\Users\\LENOVO\\Downloads\\chromedriver_win32\\chromedriver.exe'

driver = webdriver.Chrome(driver_path, chrome_options=options)
                                     
# Inicializamos el navegador
driver.get('https://www.bumeran.com.pe/empleos-area-ingenierias.html')


# Leeremos el contador de páginas
try:
    page_counter = int(driver.find_element_by_xpath('//*[@id="listado-avisos"]//button[@class="Pagination__PageLink-sc-el3mid-2 cDREQp"]').text)
except:
    print('El HTML del contador ha variado, por favor identifícalo nuevamente.')
page_counter

links = []
for i in range(0,page_counter):
    try:
        global_links = driver.find_element_by_id("listado-avisos")
        for lineas in global_links.find_elements_by_tag_name('a'):
            url = lineas.get_attribute("href")
            if '=true' not in url:
                links.append(url)
        print(f'page {i}')
        
        WebDriverWait(driver,5)\
            .until(EC.element_to_be_clickable((By.XPATH,
                                               '//*[@id="listado-avisos"]//i[@class="icon-icon-light-caret-right Icon__I-sc-1ycz6f6-0 jEQsps"]')))\
        .click()
        sleep(3)
    except:
        pass


Titulo                =  [] 
Empresa               =  []
Publicado             =  []
Departamento          =  []
Discapacidad          =  []
Horas_a_trabajar      =  []
Salario               =  []
Profecion             =  []
Modalidad             =  []
Detalles_aviso        =  []
Idioma                =  []
Experiencia           =  []
Datos_empresa         =  []
Industria             =  []
Ubicacion             =  []


Datos_emp = []
i = 1
for work in links:
    driver.get(work)
    sleep(2)

    titulo     = [driver.find_element_by_xpath('//*[@id="root"]//h1[@class="FichaAvisoSubHeader__Heading-sc-1poii24-3 jrkSCk Title__H1-sc-2yd2j1-0 dmYIZt"]').text]

    empresa    = [driver.find_element_by_xpath('//*[@id="root"]//h2[@class="FichaAvisoSubHeader__Company-sc-1poii24-2 bkFDjf"]')]

    # Atraparemos un cuadro completo para acorralar nuestra información posiblemente cambiante
    try:
        cuadro_detalles = driver.find_elements_by_xpath('//*[@id="root"]//h2[@class="DetalleAviso__InfoLabel-sc-1kfhr6j-1 kpobIa"]')

        publicado       = [cuadro_detalles[0].text]

        deprt_dist      = [cuadro_detalles[1].text]
        
        discapacidad    = ['No especificado']

        hours_work      = [cuadro_detalles[2].text]

        salario         = [cuadro_detalles[3].text]

        profes_spcial   = [cuadro_detalles[4].text]

        modalidad       = [cuadro_detalles[5].text]
        
    except:
        
        publicado           = [driver.find_element_by_xpath('//*[@id="root"]//li[1]/h2').text]
        
        try:
            deprt_dist      = [driver.find_element_by_xpath('//*[@id="root"]//div[1]/ul[1]/li[2]/a/h2').text]
        except:
            deprt_dist      = ['No especificado']
        try:   
            discapacidad    = [driver.find_element_by_xpath('//*[@id="root"]//a[contains(text(),"Apto")]').text]
        except:
            discapacidad    = ['No especificado']
            
        hours_work          = [driver.find_element_by_xpath('//*[@id="root"]//li[1]/a/h2').text]

        salario             = [driver.find_element_by_xpath('//*[@id="root"]//ul[2]/li[2]/h2').text]

        profes_spcial       = [driver.find_element_by_xpath('//*[@id="root"]//ul[2]/li[3]/a/h2').text]

        modalidad           = [driver.find_element_by_xpath('//*[@id="root"]//ul[2]/li[4]/a/h2').text]
        
        
        
        
    detalles_aviso  = [driver.find_element_by_xpath('//*[@id="root"]//div[@class="FichaAviso__DescripcionAviso-sc-b2a7hd-11 kjUXkd"]').text]

    # Extraemos información basada en palabras por etiqueta

    idioma      = []
    experiencia = []

    to_shred =  driver.find_elements_by_xpath('//*[@id="root"]//div[@class="FichaAviso__DescripcionAviso-sc-b2a7hd-11 kjUXkd"]/p')
    
    for x in to_shred:

        if 'nglés'in x.text or 'dioma' in x.text:
            idioma += [x.text]


        if 'xperiencia' in x.text or 'XPERIENCIA' in x.text:
            experiencia += [x.text]
            
    if len(idioma) == 0:
        idioma                  =            ['No especificado']
        
    if len(experiencia) == 0:
        experiencia             =            ['No especificado']

    # ¡Damos clic en información de la empresa solo si existe!
    try:
        driver.find_element_by_xpath('//*[@id="root"]//li[@class="Nav__NavItem-sc-x1nawg-1 mixins__NavTabItem-sc-n4k9uc-2 hKxcJw"]').click()
        sleep(1)

        datos_emp    = [driver.find_element_by_xpath('//*[@id="root"]//div[@class="FichaAviso__DescripcionEmpresa-sc-b2a7hd-17 jFQyIK"]').text]

        # Atraparemos un cuadro completo para acorralar nuestra información posiblemente cambiante

        info_empresa = driver.find_elements_by_xpath('//*[@id="root"]//div[@class="FichaAviso__DetalleLookAndFeel-sc-b2a7hd-12 cQVNg"]/p')

        industria    = [info_empresa[1].text]

        ubicacion    = [info_empresa[3].text]

    except:
        datos_emp    = ['Confidencial']

        industria    = ['Confidencial']

        ubicacion    = ['Confidencial']

    Titulo           +=  titulo
    Empresa          +=  empresa
    Publicado        +=  publicado
    Departamento     +=  deprt_dist
    Discapacidad     +=  discapacidad
    Horas_a_trabajar +=  hours_work
    Profecion        +=  profes_spcial
    Modalidad        +=  modalidad
    Detalles_aviso   +=  detalles_aviso
    Idioma           +=  idioma
    Experiencia      +=  experiencia
    Datos_empresa    +=  datos_emp
    Industria        +=  industria
    Ubicacion        +=  ubicacion
    print(f'page {i}')
    i += 1

    driver.quit

pd.DataFrame({ "TITULOS": Titulo,
              "EMPRESA": Empresa,
              "PUBLICADO": Publicado,
              "DEPARTAMENTO": Departamento,
              "DISCAPACIDAD": Discapacidad,
              "TIPO_CONTRATO":Horas_a_trabajar,
              "PROFECION": Profecion,
              "MODALIDAD": Modalidad,
              "DETALLES_AVISO": Detalles_aviso,
              "IDIOMA": Idioma,
              "EXPERIENCIA": Experiencia,
              "DATOS_EMPRESA": Datos_empresa,
              "INDUSTRIA": Industria,
              "UBICACION_EMPRESA":Ubicacion}).to_excel(r"ingenierias.xlsx" , index = False )
```
